```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.model_selection import cross_val_score

from sklearn.tree import DecisionTreeClassifier
from sklearn.compose import ColumnTransformer
```

```python
df = pd.read_csv('train.csv')[['Age','Fare','SibSp','Parch','Survived']] #you can use usecols = ['column1','column2] also

```

```python
df.dropna(inplace=True)
```

```python
df.head()
```

```python
df.shape
```

```python
df['Family']=df['SibSp']+df['Parch']
```

```python
df.drop(columns=['SibSp','Parch'],inplace=True)
```

```python
df.head()
```

```python
x=df.drop(columns=['Survived'])
y=df['Survived']

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
clf = DecisionTreeClassifier()
clf.fit(x_train,y_train)
y_pred = clf.predict(x_test)
accuracy_score(y_test,y_pred)
```

```python
np.mean(cross_val_score(DecisionTreeClassifier(),x,y,cv=10,scoring='accuracy'))
```

# now we will use the concept of binarization. here in the family column we will be finding if any passenger is travelling alone or not.
if alone then false (0)
else true (1)

```python
#applying binarization

from sklearn.preprocessing import Binarizer

trf = ColumnTransformer([
    ('bin',Binarizer(copy=False),['Family'])
],remainder='passthrough')
```

```python
x_train_trf = trf.fit_transform(x_train)
x_test_trf = trf.fit_transform(x_test)
```

```python
pd.DataFrame(x_train_trf,columns=['Family','Age','Fare'])
```

```python
clf = DecisionTreeClassifier()

clf.fit(x_train_trf,y_train)
y_pred1 = clf.predict(x_test_trf)
accuracy_score(y_test,y_pred1)
```

```python
x_trf = trf.fit_transform(x)
np.mean(cross_val_score(DecisionTreeClassifier(),x_trf,y,cv=10,scoring='accuracy'))
```
