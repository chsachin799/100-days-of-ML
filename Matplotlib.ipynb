```python
import matplotlib.pyplot as plt
%matplotlib inline

```

```python
import numpy as np
```

```python
x=np.arange(0,10)
y=np.arange(10,20)
```

```python
a=np.arange(40,50)
b=np.arange(50,60)
```

```python
plt.scatter(x,y,c='maroon') # here c means color and it also includes many paramters. just click shift + tab on the scatter to get the info.
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Data Plotting')
plt.savefig('Data.png')  #the figure will be saved as png .
```

```python
y=x*x
#plt.plot(x,y,'ro--')
# here 'r' for color and 'o' will be represented across the graph and the same goes for '--' . You can also try just 'r--','r0','r*-',etc.
# else replace '--' dashed with linestyle='dashed'

plt.plot(x,y,'maroon',linestyle='dashed',linewidth=2,markersize=12)
```

# Creating Subplots

```python
plt.subplot(2,2,1) # denotes total no of rows and cols and the third parameter deontes the position number.
plt.plot(x,y,'r--')
plt.subplot(2,2,2)
plt.plot(x,y,'go')
plt.subplot(2,2,3)
plt.plot(x,y,'r*-')
plt.subplot(2,2,4)
plt.plot(x,y,'g*')
```

```python
x=np.arange(0,4*np.pi,0.1)
y=np.sin(x)
plt.title('Sine Wave Form')
plt.plot(x,y)
```

```python
labels = 'Python','C++','Ruby','Java'
sizes=[215,130,245,210]
colors=['gold','yellowgreen','lightcoral','lightskyblue']
explode=(0.1,0,0,0)

plt.pie(sizes,explode=explode,labels=labels,colors=colors,autopct='%1.1f%%',shadow=True)
plt.axis('equal')
plt.show()
```
