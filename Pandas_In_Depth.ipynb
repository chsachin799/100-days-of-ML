## #Find the most destructive death over batsman in the history of IPL

1.Strike Rate = (Number of runs/No.of balls)*100
2.Min batted 200 balls in over 16-20


```python
import pandas as pd
delivery = pd.read_csv('deliveries.csv')
```

```python
delivery
```

```python
delivery.groupby('batter')['batsman_runs'].count()
```

```python
death_overs = delivery[delivery['over']>15]
```

```python
death_overs.groupby('batter')['batsman_runs'].count()
```

```python
all_batsman=death_overs.groupby('batter')['batsman_runs'].count()
x=all_batsman>200
```

```python
batsman_list=all_batsman[x].index.tolist()
```

```python
batsman_list
```

```python
#runs scored by all these 71 batsman
#balls played by these 71 batsman

final = delivery[delivery['batter'].isin(batsman_list)]
final
```

```python
# now from these final result we need to calculate the runs scored and the balls played

runs = final.groupby('batter')['batsman_runs'].sum()
```

```python
balls = final.groupby('batter')['batsman_runs'].count()
```

```python
sr=(runs/balls)*100
```

```python
sr
```

```python
sr.max()
```

```python
sr['MS Dhoni']
```

# The merge function

find in particular season which batsman had the orange cap and purple cap.

but in matches.csv file we see that the datas of season are only present.
While in deliveries.csv we can see the seasons are not present but the batsman and bowlers data are present.
In this case we need to merge both the dataframes into single one so that we can work on it.


```python
# for merge we need to have a similar column
# we have 'id' in matches.csv and 'match_id' in deliveries.csv
new = delivery.merge(data,left_on='match_id',right_on='id')
```

```python
print(delivery.shape)
print(data.shape)
print(new.shape)
```

```python
new.groupby(['season','batter'])['batsman_runs'].sum().sort_values(ascending=False).reset_index().drop_duplicates(subset='season',keep='first').sort_values('season')
 # we used reset_index to convert the series into the dataframe
 # here in each season we will find out the numbers of runs the batmans has made.
```

```python
new=new.groupby(['season','batter'])['batsman_runs'].sum().sort_values(ascending=False).reset_index().drop_duplicates(subset='season',keep='first').sort_values('season')[['season','batter']]
```

```python
new
```

```python
#rename the columns
new.rename(columns={'season':'SEASON','batter':'PLAYER'})
```
