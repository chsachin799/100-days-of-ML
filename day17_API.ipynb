```python
df = pd.DataFrame()

for i in range(1, 525):
    response = requests.get('https://api.themoviedb.org/3/movie/top_rated?api_key=d1cd885f1979e575210a82b8b073774d&language=en-US&page={}'.format(i))
    data = response.json()

    if 'results' not in data:
        continue

    temp_df = pd.DataFrame(data['results'])[['id','title','overview','release_date','popularity','vote_average','vote_count']]
    df = pd.concat([df, temp_df], ignore_index=True)

```
