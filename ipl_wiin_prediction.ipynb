```python
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings("ignore")  #it ignores or removes warnings

from sklearn.model_selection import train_test_split
```

# First Dataset

```python
first_dataset = pd.read_csv('dataset.csv')
first_dataset.head()
```

# loading second dataset

```python
second_dataset = pd.read_csv('IPL Matches 2008-2020.csv')
second_dataset.head()
```

# performing data cleaning

### #selecting the only first innings to find the target of each match


```python
total_runs = first_dataset.groupby(['match_id','inning']).sum()['total_runs'].add(1).reset_index()
```

```python
total_runs=total_runs[total_runs['inning']==1]
```

```python
total_runs
```

```python
match_df = second_dataset.merge(total_runs[['match_id','total_runs']],left_on='id',right_on='match_id')
```

```python
match_df
```

```python
match_df = match_df[['id','city','winner','total_runs']]
```

```python
match_df=match_df.rename(columns={'id':'match_id'})
```

```python
second_innings = match_df.merge(first_dataset,on='match_id')
second_innings
```

```python
second_innings=second_innings[second_innings['inning']==2]
second_innings.shape
```

# removing repeating and old teams

```python
second_innings['batting_team'].value_counts()
```

```python
second_innings['batting_team'] = second_innings['batting_team'].replace('Delhi Daredevils','Delhi Capitals')
second_innings['bowling_team'] = second_innings['bowling_team'].replace('Delhi Daredevils','Delhi Capitals')

second_innings['batting_team'] = second_innings['batting_team'].replace('Deccan Chargers','Sunrisers Hyderabad')
second_innings['bowling_team'] = second_innings['bowling_team'].replace('Deccan Chargers','Sunrisers Hyderabad')
```

```python
second_innings['batting_team'].value_counts()
```

```python
second_innings['bowling_team'].value_counts()
```

```python
teams = [
    'Kolkata Knight Riders',
    'Chennai Super Kings',
    'Rajasthan Royals',
    'Mumbai Indians',
    'Kings XI Punjab',
    'Royal Challengers Bangalore',
    'Delhi Capitals',
    'Sunrisers Hyderabad'
]
```

```python
second_innings = second_innings[second_innings['batting_team'].isin(teams)]
second_innings = second_innings[second_innings['bowling_team'].isin(teams)]
```

```python
# 1. Convert the column to numeric (errors='coerce' turns non-numeric values into NaN)
second_innings['total_runs_y'] = pd.to_numeric(second_innings['total_runs_y'], errors='coerce')

# 2. Fill any NaNs with 0 (optional but recommended for cumsum)
second_innings['total_runs_y'] = second_innings['total_runs_y'].fillna(0)

# 3. Perform the groupby cumsum (selecting the column BEFORE the cumsum)
second_innings['current_Score'] = second_innings.groupby('match_id')['total_runs_y'].cumsum()
```

```python
second_innings
```

```python
second_innings['runs_left'] = second_innings['total_runs_x'] - second_innings['current_Score']
second_innings
```

```python
second_innings['balls_left'] = 126-(second_innings['over']*6 + second_innings['ball'])
second_innings
```

# Getting wickets

```python
#filling null values with zero
second_innings['player_dismissed']=second_innings['player_dismissed'].fillna("0")

#if column value is not zero mark it as 1
second_innings['player_dismissed']=second_innings['player_dismissed'].apply(lambda x :x if x=="0" else "1")

#changing datatype pf column name
second_innings['player_dismissed']=second_innings['player_dismissed'].astype('int')

#finding wickets using cumulative sum function
# Select the column first, then apply cumsum
wickets = second_innings.groupby('match_id')['player_dismissed'].cumsum()

#calculating wickets left by simple maths
second_innings['wickets_remaining']=10-wickets
second_innings
```

# Current Run Rate

```python
second_innings['crr'] = round((second_innings['current_Score']*6)/(120-second_innings['balls_left']),2)
second_innings
```

# Required Run rate

```python
second_innings['rrr'] = round((second_innings['runs_left']*6)/(second_innings['balls_left']),2)
```

# Winner

```python
def winner(team):
  return 1 if team['batting_team'] == team['winner'] else 0;
```

```python
second_innings['results'] = second_innings.apply(winner,axis=1)
second_innings
```

# only taking required columns

```python
required_data = second_innings[['batting_team','bowling_team','city','runs_left','balls_left','wickets_remaining','total_runs_x','crr','rrr','results']]
required_data
```

## Sampling Data

```python
final_data = required_data.sample(required_data.shape[0])
#it re-shuffles the data
```

# Dropping nuii values

```python
final_data.dropna(inplace=True)
```

# remove balls where balls_left ==0 to remove -inf and inf from rrr column

```python
final_data = final_data[final_data['balls_left']!=0]
```

# Storing data in new csv file

```python
final_data.to_csv('final_data.csv',index=False)
```

# Building Model

```python
from sklearn.model_selection import train_test_split as tts
x=final_data.iloc[:,:-1]
y=final_data.iloc[:,-1]
```

```python
x_train,x_test,y_train,y_test = tts(x,y,test_size=0.25,random_state=100)
x_train
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

trf = ColumnTransformer([
    ('trf',OneHotEncoder(sparse_output=False,drop='first'),['batting_team','bowling_team','city'])
],remainder='passthrough')
```

```python
from sklearn import pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression


# 1. Identify your categorical and numerical columns
categorical_features = ['batting_team', 'bowling_team', 'city']
numerical_features = ['runs_left', 'balls_left', 'wickets_remaining', 'crr', 'rrr']

# 2. Create a preprocessor that handles both types of data
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numerical_features), # Scales numbers for better accuracy
        ('cat', OneHotEncoder(sparse_output=False), categorical_features) # Converts text to 0s and 1s
    ])

# 3. Re-run your model comparison with the Pipeline

pipe = Pipeline([
    ('step1',trf),
    ('step2',LogisticRegression(solver='liblinear'))
])
```

```python
pipe.fit(x_train,y_train)
```

```python
from sklearn.metrics import accuracy_score
y_pred = pipe.predict(x_test)
accuracy_score(y_test,y_pred)
```
